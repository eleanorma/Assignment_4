```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
url = "https://www.showstart.com/event/list?pageNo=1&pageSize=20&cityCode=10&showStyle=26&sortType=9"
html = urlopen(url)
soup = BeautifulSoup(html,'html.parser')
print(soup)
```

```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
import pandas as pd

performances = []
artists = []
prices = []
times = []
addresses = []

for item in soup.find_all("a", attrs={"class":"show-item item"}):
    performance = item.find("div",attrs={"class":"title"})
    artist = item.find("div",attrs={"class":"artist"})
    price = item.find("div",attrs={"class":"price"}) 
    time = item.find("div",attrs={"class":"time"}) 
    address = item.find("div",attrs={"class":"addr"}) 
    performances.append(performance.text)
    artists.append(artist.text)
    prices.append(price.text)
    times.append(time.text)
    addresses.append(address.text)

df=pd.DataFrame({"Event Title":performances,"Artists":artists,"Price":prices,"Time":times,"Venue":addresses}) 
df.to_csv('liveconcerts.csv')
df  
```

```python
#data cleaning
df = pd.read_csv("liveconcerts.csv",index_col=[0])
df['Artists'] = df['Artists'].str.replace('艺人：', '')
df['Price'] = df['Price'].str.replace('价格：', '')
df['Time'] = df['Time'].str.replace('时间：', '')
df
```
